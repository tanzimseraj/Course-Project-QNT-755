```python
#Import Process
import csv
import numpy as np

# Parse CSV File into a List
csvFile = open("C:/Users/tanzi/Downloads/WHO-COVID-19-global-data/WHO-COVID-19-global-data.csv")
csvReader = csv.reader(csvFile)

# Change CSV List in a tab-separated String
tsvString = ("\t".join(row) for row in csvReader)

# Now, numpy won't have any issues with commas in the middle of the data
data = np.genfromtxt(tsvString, delimiter="\t", dtype=None, encoding=None, names=True)

# Sort and reverse data, to get Descending order
data.sort(order=['Cumulative_deaths'])
#data = data[::-1]

print(data[['Country', 'Cumulative_deaths']])

# clean up
csvFile.close()
```

```python
##Over here  wanted to show the top ten countries with the most deaths before the outbreak
import calendar
criteria = lambda x: (x["Country"], x["Cumulative_deaths"], x["ï»¿Date_reported"])

result = list(map(criteria, data))
result = np.array(result, dtype=[("Country", "U25"), ("Cumulative_deaths", "i8"), ("Date_reported", "datetime64[D]")])
# print(len(result))
#print(result)
#print(len(result))

# 'Outbreak' date range
where_outbreak_dates = np.logical_and(result["Date_reported"] np.datetime64 ==('2020-06-22'))

# Convert to datetime objects
result_as_date_times: np.ndarray = result["Date_reported"].astype(object)

# Filter to last day in month
is_last_day_in_month = lambda x: calendar.monthrange(x.year, x.month)[1] == x.day
where_last_day_in_month = np.array(list(map(is_last_day_in_month, result_as_date_times)))

new_result = result[
    np.logical_and(where_outbreak_dates, where_last_day_in_month)
]

print(new_result)
```
